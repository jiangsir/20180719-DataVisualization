```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(25,15))

dfrf = pd.read_csv('2017_rainfall.csv')
dfpm25 = pd.read_csv('2017_pm25.csv')
dfpm10 = pd.read_csv('2017_pm10.csv')

dfrf['summ'] = dfrf.sum(axis=1)
dfpm25['mean'] = dfpm25.mean(axis=1)
dfpm10['mean'] = dfpm10.mean(axis=1)

dfrf['summ'].plot()
dfpm10['mean'].plot()
dfpm25['mean'].plot()
# 橘色 pm2.5
# 綠色 pm10
dfrf.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

plt.figure(figsize=(25,15))

dfrf = pd.read_csv('2017_rainfall.csv')
dfpm25 = pd.read_csv('2017_pm25.csv')
dfpm10 = pd.read_csv('2017_pm10.csv')


startdate = '2017/04/01'
enddate = '2017/10/01'
dfrf = dfrf.loc[(dfrf['日期'] > startdate) & (dfrf['日期'] <= enddate)]
dfpm25 = dfpm25.loc[(dfpm25['日期'] > startdate) & (dfpm25['日期'] <= enddate)]
dfpm10 = dfpm10.loc[(dfpm10['日期'] > startdate) & (dfpm10['日期'] <= enddate)]

dfrf['mean'] = dfrf.mean(axis=1)
dfpm25['mean'] = dfpm25.mean(axis=1)
dfpm10['mean'] = dfpm10.mean(axis=1)


dfrf['mean'].plot()
dfpm10['mean'].plot()
dfpm25['mean'].plot()
# 橘色 pm2.5
# 綠色 pm10
dfrf.head()
```
